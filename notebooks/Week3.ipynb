## 1.4 Overview of Delay in Packet-Switched Networks

  *Recall*: that a packet starts at a source host, and passes through a series of routers, and ends its journey at the destination. The packet travels from one Node (host or router) to the subsequent node (another host or router) along this path.

  The packet can suffer from several types of delays at *each* node along the path. The most important of these delays are the **nodal processing delay**, **queuing delay**, **transmission delay**, and **propagation delay**. Aggregating these will give the **total nodal delay**. The best performance depends on these. 


## *Types of Delay*:

  A part of its end-to-end route beween source and destination, a packet is ent from the upstream node through router A to router B. Our goal is to characterize the *nodal delay* at router A. note that router A has an outbound link leading to router B. This link is preceded by a queue (buffer). When the packet arrives at router A from the upstream node, router A examines the packet's header to determine the appropriate outbound link for the packet and then directs the packet to this link. 

  In this example, the outbound link for the packet is the one that leads to router B. It is this transmitted on a link only if there is no other packets currently being transmitted on the link and if there are no other packets preceding it in the queue. If the link is currently busy or if there are other packets already queued for the link, the newly arriving packet will then join the queue.

<img src='./screenshots/sources-of-packet-delay.png' width='300' height='300'>

### *Processing Delay*: 

- check bit errors
- determine output link
- typically < msec

  The time required to examine the packet's header and determine where to direct the packet is part of the ***processing delay***. The processing delay can also indlude other factors, such as the time nedded to check for bit-level errors in the packet that occurred in transmitting the packet's bits from the upstream node to router A. Processing delays in high-speed routers are typically on the order of microseconds or less. After the nodal prcoess, the router directs the packet to the queue that preceeds the link to router B.

### *Transmission Delay*: 

- *L*: packet length (bits)
- *R*: link *bandwidth* (bps)
- transmission delay and propagation delay are not the same

  Assuming that packets are transmitted in a first-come-first-served manner, as is common in packet-swtched networks, our packet can be transmitted only after all the packets that have arrived before it have been transmitted. Denote the length of the packet by *L* bits, and denote the transmission rate of the link from router A to router B by R bps.
  
For example: a 10 Mbps Ethernet link, the rate is R=10Mbps, and for 100 Mbps Ethernet link, the rate is R=100Mbps. The ***transmission delay*** is $L/R$. This is the amount of time required to push (transit) all of the packet's bits into the link. These are typically on the order of microseconds to milliseconds.

### *Propagation Delay*:

- *d*: length of physical link
- *s*: propagation speed $ (~2 * 10^8 m/s $
- where the *propagation of d* is $ d/s $

  When a bit is pushed into the link, it needs to propagate to router B. The time required to propagate from the beginning of the link to router B is ***propagation delay***. The bit propagates at the propagation speed of the link. The propagation speed depends on the physical medium of the link, and is in the range of:
$2 * 10^8 m/s$ to $3 * 10^8 m/s$ and is equal to, or a little less than, the speed of light. The propagation delay is the distance between two routers divided by the propagation speed. That is, the propagation delay is $d/s$, where *d* is the distance between router A and router B and *s* is the propagation speed of the link. Once the last bit of the packet propagates to node B, it and *all* the preceding bits of the packet are stored in router B. The process with repeat at each router while performing forwarding. Propagation delays are on the order of milliseconds.


### *Queueing Delay*:

- time waiting at output link for transmission
- depends on congestion level of router
- *R*: link bandwidth (bps)
- *L*: packet length (bits)
- *a*: average packet arrival rate
- $ La/R = 0 $: avg. queueing delay small
- $ La/R \geq 1 $: avg. queueing delay large, more "work" arriving


Unlike the other three delays; this delay can vary from packet to packet. For example, if 10 packets arrive at the same time, the first will transmit without any queueing delay, while the last packet transmitted will suffer a relatively large queueing delay. To gain insight, let a denote the average rate at which packets arrive at the queue (a is in units of packets/sec). Recall that *R* is the transmission rate; that is, it is the rate (in bits/sec) at which bits are pushed out of the queue. The ratio $ La/R $, called the ***traffic intensity***, and it often plays an important role in estimating the extent of the queueing delay. If the product is greater than 1, then the average rate at whichi bits arrive at the queue exceeds the rate at whichc the bits can be transmitted ffrom the queue. Unfortunately under these conditions the delay will eventually reach infinite. *Design your system so that the traffic intensity is no greater than 1*.

Now consider the case $La/R \leq 1$. The nature of the arriving traffic impacts the queueing delay. If the packets arrive periodically - that is, one packet arrives every $ L/R $ seconds - then every packet will arrive at an empty queue and there will be no queueing delay. However, suppose *N* packets arrive simultaneously every $(L/R)N$ seconds. Then the first packet transmitted has no queueing delay; the second packet transmitted has a queueing delay of $ L/R $ seconds. Inductively, the *n*th packet transmitted has a queueing delay of $ (n-1)L/R $ seconds. 

Typically the arrival process to a queue is *random*; that is, the arrivals do not need to follow any patterns and the packets are spaced apart by random amounts of time. In general the quantity $ La/R $ is not usually sufficient to fully characterize the queueing delay statistics. If the average queueing delay is close to 1, there will be intervals of time when the arrival rate exceeds the transmission capacity due to variations in packet arrival rate), and a queue will form during these periods of time; when arrival rate is less than the transmission capacity, the length of the queue will shrink, but as it approaches 1, the average length gets longer and longer.

#### Packet Loss: 

In reality, a queue preceding a link has a finite capacity, although the queueing capacity greatly depends on the router design and cost. Typically if a packet arrives to find a full queue with no place to be stored, the router will *drop* that packet, and it will be *lost*. From an end-system viewpoint, a packet loss will look like a packet having been transmitted into the network core but never emerging from the network at the destination. The fraction of lost packets increases as the traffic intensity increases and the node is often measured not only in terms of delay but in terms of the probability of packet loss. A lost packet may be retransmitted on an end-to-end basis in order to ensure that all data are eventually transferred from source to destination. 

### *End-to-End Delay*: 

Considering the total delay from source to destination. Suppose there are N-1 routers between the source host and the destination host. For a moment the network is uncongested, the process delay at each router and at the source host is d_{prop}, the transmission rate out of each router and out of the source host is *R* bits/sec, and the propagation on each link is d_{prop}. The nodal delays accumulate and give an end-to-end delay:
> $ d_{end-to-end} = N * (d_{proc} + d_{trans} + d_{prop}) $ 

$ d_{trans} = L/R $, where *L* is the packet size.



#### "Real" Internet Delays and Routes:
**Traceroute** programs provide delay measurements from source to router along end-to-end internet path towards destination.

For all *i*:
- send three packets that will reach router *i* on path towards destination
- router *i* will return packets to sender
- sender times interval between transmission and reply.

Suppose there are N-1 routers between the source and the destination. Then the source will send *N* special packets into the network, with each packet addressed to the ultimate destination. These *N* special packets are marked *!* through *N*, with the first packet marked *1* and the last packet marked *N*. When the *n*th router receives the *n*th packet marked *n*, the router does not forward the packet toward its destination, but instead sends a message back to the source. When the destination receives the *N*th packet, it too returns a message back to the source. The source will record the time that elapses between when it sends a pecket and when it reveices the corresponding return message; it also records the naem and aaddress of the router that returns the message. The source can reconstructed the route taken by packets flowing from source to the destination. The source can determine the round-trop delays to all the intercening routers. Traceroute actually repeats the experiment just described three times. So the source actually sends $3 \textbullet N$ packets to the destination.

### *Throughput in Computer Networks*

In addition to delay and packet loss, another critical performance measure in computer networks is end-to-end throughput. To define throughput, consider transferring a large file from Host A to Host B across a computer network. This could be a video from one peer to another in a P2P file sharing system. The **instantaneous throughput** at any instant of time is the rate (in bits/sec) at which Host B is receiging the file. If the file consists of *F* bits and the transfer is *F/T* bits/sec. It is desirable to have a low delay and an instantaneous throughput consistently above some threshold. 

<img src='./screenshots/throughput-transfer.png' width='300' height='300'>

Let $R_{s}$ denote the rate of the link between the serer and the router; and $R_{c}$ denote the rate of the link between the router and the client. 

Suppose that the only bnits being sent in the entire network are those from the server to the client. In this ideal scenaria, what is ther server-to-client throughput? To answer this question, we may think of bits as *fluid* and communication links as *pipes*. The server cannot pump bits through its link at a rate faster than $R_{s}$ bps; and the router connot forward bits at a rate faster than $R_{c}$ bps. 
- If $R_{s}$ < $R_{c}$, then the bits pumped by the server will "flow" right through the router and arrive at the client at a rate of $R_{s}$ bps, giving a throughput of $R_{s}$. 
- If, on the other hand, $R_{c}$ < $R_{s}$, then the router will not be able to forward bits as quickly as it receives them. 

In this case, bits will only leave the router at rate $R_{c}$, giving an end-to-end throuhgtput of $R_{c}$. Note, if bits continue to arrive at the router at rate $R_{s}$, and coninue to leave the router at $R_{c}$, the backlog of bits at the router wating for transmission to the client will grow exponentially. Thus, for this simple two-link network, the throughput is min{$R_{c}$, $R_{s}$}, that is, it is the transmission rate of the **bottleneck link**. 

Having determined the throughput, the approximate time it takes to transfer a large file of *F* bits from server to client as *F*/min{$R_{c}$, $R_{s}$}. 

Suppose, a MP3 file is being downloaded of *F*=32 million bits, and the server is tranmitting them at a rate of $R_{s}$=2 Mbps, and the access link can receive them at a rate of $R_{c}$=1 Mbps. The time needed to transfer the file is then 32 seconds. Of course, these expressions for throughput and transer time are ony approximations, as they do not account for store-and-forward and processing delays as well as protocol issues.

The figure above shows a network with *N* links between the server and the client, with the transmission rates of the *N* link being $R_{1}, R_{2}, ..., R_{N}$. Applying the same analysis as for the two-link network, the throughput for a file transfer from server to client is min{$R_{1}, R_{2}, ..., R_{N}$}, which is once again the transmission rate of the bottleneck link along the path between server and client.
