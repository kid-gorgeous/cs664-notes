#### Hypertext Transfer Protocol:
This is the webs *application layer protocol*, and the heart of the Web. It is defined in [RFC 1945](https://www.rfc-editor.org/rfc/rfc1945), and [RFC 2616](https://www.rfc-editor.org/rfc/rfc2616). HTTP is implemented in two programs: client, and server. Tese programs exist of two different systems, and will communicate with the exchange of HTTP messages. HTTP will defines the structure of these messages and how the client and server exchange the messages. 

#### Terminology: 
- Webpages: consists of objects
- Objects: is a simple file type, that is addressable by a single URL.
- base HTML file: will contain HTML, and the files included within it.
- URL: has two components: the hostname of the server that houses the object, and the objects path name.
- Web browser: implementation of client side HTTP
- Web server: implementation of server side of HTTP request messages

HTTP will use TCP (Transmission Control Protocol) transport protocol. The HTTP client will make the first initial TCP connection with the server. Once a connection is established, the server processes access TCP through their socket interfaces. Once the client sends a message into its socket interface, the message is now "in the hands" of TCP.  

HTTP is siad to be a **stateless protocol** because there is no information about the client (unless through Cache), and uses the client-server. 

### 2.2.2 - Non-Persistent and Persistent Connections

Time is very important, with client making a series of requests and the server responding to each of the requests. Requests can be back-to-back, periodically, or at regular intervals, or intermittently. Client-server interactions is taking place over TCP, but when there is multiple requests be sent over the same TCP connections, or should all of the requests and their corresponding responses be sent over the *same* TCP connections. In the former approach, the application is said to use ***non-persistent connections***, and the latter approach ***persistent connection***.
