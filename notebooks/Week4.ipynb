#### Packet Switching, Store and forward mechanism:

- **Hosts**: sending function
    - takes a message, breaks into smaller chunks and into ***packets*** (of length L)
    - transmits packets into access network at **transmission rate R**
        - link transmission rate, aka link ***capacity, aka link bandwidth***
      
packet transmission delay = time needed to transmit *L*-bit packet into link = *L*(bits)/*R*(bits/sec)

#### Store-and-Forward:

- The tranmission rate will be determined by $\frac{L}{R}$ seconds to transmit (push out) *L*-bit packet into link at *R* bps.

- **Store and Forward**: the entire packet must be received at one router location before it will begin the forwarding operation to the next.
    - there will be an end to end delay = $\frac{2L}{R}$ (assuming zero propagation delay)
    - Queuing and Data Loss is caused by high transmission frequencies over time
        - packets queue, wait to be transmitted
        - packets are dropped if memory buffer fills

#### Throughput:

Considering transferring a large file from Host A to Host B across a computer network. This transfer might be, a video from peer to peer to another in a P2P file sharing system. The **instantaneous throughput** at any instant of time is the rate (in bits/sec) at which Host B is receiving the file.

If a file consists of *F* bits and transfer takes *T* seconds for Host B to receive all *F* bits, then **average throughput** of the file. Transfer is *F/T* bits/sec.
